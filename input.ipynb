```python
movies = movies_df.merge(credits_df,on="title")
```

```python
pd.set_option('display.max_columns', None)
movies.head()
```

```python
movies = movies.drop('movie_id', axis=1)
```

```python
movies.shape
```

```python
movies.dtypes
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=5:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['production_companies'] = movies['production_companies'].apply(convert3)
movies['production_countries'] = movies['production_countries'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```
